1.- Obtener la norma 1, 2, ∞.

De los vectores:

a) V1 = (1, 0, 2)

```python
import math as m

def norma(p, v):
    result = 0.0
    if p == 1:
        for i in range(0,len(v)):
            result += abs(v[i])
        return(result)
    elif p == 2:
        for i in range(0,len(v)):
            result += abs(v[i])**2
        return (m.sqrt(result))
    elif p == "∞":
        for i in range (0,len(v)):
            v[i] = abs(v[i])
        return(max(v))
```

```python
v1 = [1,0,2]

print ("norma 1: %f" %norma(1,v1))
print ("norma 2: %f" %norma(2,v1))
print ("norma ∞: %f" %norma("∞",v1))
```

b) V2 = (-6, 5)

```python
v2 = [-6,5]

print ("norma 1: %f" %norma(1,v2))
print ("norma 2: %f" %norma(2,v2))
print ("norma ∞: %f" %norma("∞",v2))
```

c) V3 = (√2,-1,0,1)

```python
v3 = [m.sqrt(2),-1,0,1]

print ("norma 1: %f" %norma(1,v3))
print ("norma 2: %f" %norma(2,v3))
print ("norma ∞: %f" %norma("∞",v3))
```
